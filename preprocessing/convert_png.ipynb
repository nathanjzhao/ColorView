```python
import os
from PIL import Image

# Define the source directory
src_dir = '../data/Military/Train/ACD1K_CAMO/GT_Object'  # replace with your directory
```

```python
# Get a list of all jpeg files in the source directory
files = [f for f in os.listdir(src_dir) if f.lower().endswith(('.jpg', '.jpeg'))]

# Process each file
for file in files:
    # Open the image file
    img = Image.open(os.path.join(src_dir, file))
    # Convert the image to png
    img.save(os.path.join(src_dir, os.path.splitext(file)[0] + '.png'))

# Process each file
for file in files:
    # Delete the file
    os.remove(os.path.join(src_dir, file))
```
