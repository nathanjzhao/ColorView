```python
import cv2
import os
import glob
import numpy as np

# Define the source directory
src_dir = '../data/Military/Train/ACD1K_CAMO/Image'

# Get a list of all JPG files in the source directory
jpg_files = glob.glob(os.path.join(src_dir, '*.jpg'))

print(jpg_files)

max_width = 0
max_height = 0

psum = np.array([0.0, 0.0, 0.0])
psum_sq = np.array([0.0, 0.0, 0.0])
count = 0

# Process each JPG file
for jpg_file in jpg_files:
    # Read the image
    img = cv2.imread(jpg_file).astype("int")
    h, w, c = img.shape
    max_width = max(max_width, w)
    max_height = max(max_height, h)
    psum += img.sum(axis=(0, 1))
    psum_sq += (img**2).sum(axis=(0, 1))
    count += w * h


total_mean = psum / count
total_std = np.sqrt((psum_sq / count) - (total_mean**2))

print("Mean:", total_mean)
print("STD:", total_std)

print("Max Width:", max_width)
print("Max Height:", max_height)
```
